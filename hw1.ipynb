__Задание 1.__ Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

*Решение:*

$f_{4}(x)$ можно линейно выразить через другие функции:

$f_{4}(x) = f_{3}(x) - f_{2}(x) - f_{1}(x) = x + 1 - 1 - e^{x} = x-e^{x}$

Следовательно, перед нами пример линейной зависимости

__Задание 2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}$$

*Решение:*

$f_{4}(x)$ можно линейно выразить через другие функции:

$f_{4}(x) = f_{3}(x) + 2f_{2}(x) + \frac{1}{2}f_{1}(x) = x^{2} + 2x + 1 = (x+1)^{2}$

__Задание 3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

*Решение:*

$ x_{1} = 2 \times 2 = 4$

$ x_{2} = 3 \times 1 = 3$

$ x_{0} = 5 \times 10 = 50$

Ответ: $x = (50, 4, 3)$

   б) в базисе $x^{2}$, $x-1$, $1$.

а) Координаты вектора - это коэффициенты перед базисными векторами

б) $3x^{2}-2x+2 = 3x^{2} - 2 \times (x - 1) + 0$

Ответ: $x = (3, -2, 0)$
